```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
           
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('dsp_latest_price_max.csv')
data = data[data.sid != 'UnKnOwN']

data = data.drop(['date'], axis=1)
data = data.sample(frac=1)
```

```python
data = data[data['sum(impressions)']>1000]
```

```python
data
```

```python
X_pre = data.drop(['max_price','sum(payout)','sum(impressions)','sum(spend)'], axis=1)
le = LabelEncoder()
X_pre['country']= le.fit_transform(X_pre['country']) 
X_pre['os']= le.fit_transform(X_pre['os']) 
X_pre['osv']= le.fit_transform(X_pre['osv']) 
X_pre['carrier']= le.fit_transform(X_pre['carrier']) 
X_pre['app_name']= le.fit_transform(X_pre['app_name']) 
X = X_pre.drop(['app_name','hour','osv','carrier'], axis=1)
y = data['max_price']
```

```python
data
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train) #training the algorithm
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test,y_pred)
```

```python
y_pred
```

```python
from sklearn.ensemble  import RandomForestRegressor
```

```python
rf_reg = RandomForestRegressor()
rf_reg.fit(X_train, y_train)
```

```python
predictions = rf_reg.predict(X_test)
```

```python
rf_reg.score(X_test,y_test)
```

```python
r2_score(y_test,predictions)
```
