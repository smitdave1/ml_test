```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
           
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('payout_list.csv')

data = data.sample(frac=1)
```

```python
data
```

```python
X_pre = data.drop(['Impressions','Time'], axis=1)
le = LabelEncoder()
X_pre['country']= le.fit_transform(X_pre['country']) 
X_pre['os']= le.fit_transform(X_pre['os']) 
X_pre['osv']= le.fit_transform(X_pre['osv']) 
X_pre['carrier']= le.fit_transform(X_pre['carrier']) 
X_pre['app_name']= le.fit_transform(X_pre['app_name']) 
X = X_pre.drop(['app_name','hour','osv','carrier'], axis=1)
y = data['max_price']*0.04
```

```python
X = data.drop(['Impressions','Time'], axis=1)
le = LabelEncoder()
X['App name']= le.fit_transform(X['App name']) 


#X['Time']= le.fit_transform(X['Time']) 

y = data['payout']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train) #training the algorithm
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test,y_pred)
```
